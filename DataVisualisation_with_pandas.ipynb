```python
data = pd.read_csv("countries.csv")
```

```python
#to get th unique values
```

```python
list (set(data.country))  
```

```python
#comparing the population growth of United States,India,Germany and China
```

```python
china =data[data.country=='China']
china
```

```python
us= data[data.country=='United States']
```

```python
us
```

```python
SA = data[data.country=='South Africa']
SA
```

```python
Germany =data[data.country=='Germany']
Germany
```

```python
india =data[data.country=='India']
india
```

```python
plt.plot (us.year,us.population / 10**6 , 'o')
plt.plot(china.year,china.population /10**6 ,'+')
plt.plot(india.year,india.population /10**6 , '--')
plt.plot(Germany.year,Germany.population /10**6 , '*')
plt.plot(SA.year,SA.population /10**6, '.')
plt.xlabel ("Year")
plt.ylabel ("Population Growth")
plt.title("Population Growth Comparision")
plt.legend (["US","China" ,"India","Germany","South Africa"])



plt.savefig('line_plot.pdf',bbox_inches="tight")  
```
